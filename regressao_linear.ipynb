```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

x = np.array([2,3,5,7,10,12,14,18])
y = np.array([4,5,7,8,11,13,15,18])

declive, interceptacao, valor_r, valor_p, desvio_padr = linregress(x, y)

plt.scatter(x,y, label = "dados")
plt.plot(x, declive * x + interceptacao, color="pink", label = "Regressao linear")
plt.legend()
plt.show();

print("coeficiente de inclinacao: ", declive)
print("interceptacao: ", interceptacao)
print("valor r: ", valor_r)
print("valor p: ", valor_p)
print("desvio padrao: ")
```

```python
import numpy as np
from scipy.stats import linregress
import matplotlib.pyplot as plt

x = np.array([2,3,5,7,10,12])
y = np.array([4,5,7,8,11,13])

novo_x = 8

inclinacao, interceptacao, _,_,_ = linregress(x,y)

predicao_y = inclinacao * novo_x + interceptacao
print("previsao para x: ", novo_x, predicao_y)

plt.scatter(x,y, label = "dados")
plt.plot(x, inclinacao * x + interceptacao, color="pink", label = "Regressao linear")
plt.legend()
plt.show();
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

x = np.array([[2,3],[3,4],[5,6],[7,8],[10,11], [12,13]])
y = np.array([5,6,8,9,12,14])

modelo = LinearRegression()
modelo.fit(x,y)

coeficientes = modelo.coef_
interceptacao = modelo.intercept_

print("Coeficientes:", coeficientes)
print("Interceptacao: ", interceptacao)

plt.scatter(x[:, 0], y,color="blue", label = "rg Arozz")
plt.scatter(x[:, 1], y, color="red", label = "rg Feijao")

plt.plot(x[:,0], modelo.predict(x), color="blue", label="regressao linear arroz")
plt.plot(x[:,1], modelo.predict(x), color="red", label="regressao linear feijao")

plt.xlabel("meses do ano")
plt.ylabel("toneladas")

plt.legend()

plt.show()
```
